```python
import pandas as pd
```

```python
data = pd.read_csv(r"..\2. eda_n_dataCleaning\cleaned_data.csv")
data.head(3)
```

```python
data['q1_num_char'] = data['question1'].str.len()
data['q2_num_char'] = data['question2'].str.len()
data.head()
```

```python
data['q1_num_words'] = data['question1'].astype(str).apply(lambda x: len(x.split(' ')))
data['q2_num_words'] = data['question2'].astype(str).apply(lambda x: len(x.split(' ')))
data.head()
```

```python
def find_common_words(row):
    common_words=[]
    for q1_word in row.astype(str)['question1'].split(' '):
        if str(q1_word).strip().lower() in str(row['question2']).strip().lower():
            common_words.append(str(q1_word).strip().lower())
    return len(set(common_words))
```

```python
data['common_word_count'] = data.apply(find_common_words, axis=1)
data.head()
```

```python
def total_words(row):
    w1 = set(row.astype(str)['question1'].strip().lower().split(' '))
    w2 = set(row.astype(str)['question2'].strip().lower().split(' '))
    return len(w1) + len(w2)
```

```python
data['total_word_count'] = data.apply(total_words, axis=1)
data.head()
```

```python
data['word_share'] = round(data['common_word_count']/data['total_word_count'], 2)
data.head()
```

```python
data.to_csv(r'basic_featured_data.csv', index=False)
```
