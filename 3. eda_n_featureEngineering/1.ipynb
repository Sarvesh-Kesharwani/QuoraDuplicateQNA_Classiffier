```python
import pandas as pd
```

```python
data = pd.read_csv(r"../1. data_gathering\data\train.csv\train.csv")
data.head(3)
```

```python
data['q1_num_char'] = data['question1'].str.len()
data['q2_num_char'] = data['question2'].str.len()
data.head()
```

```python
data['q1_num_words'] = data['question1'].astype(str).apply(lambda x: len(x.split(' ')))
data['q2_num_words'] = data['question2'].astype(str).apply(lambda x: len(x.split(' ')))
data.head()
```

```python
def find_common_words(row):
    common_words_count=0
    # return row['question1']
    for q1_word in row['question1'].split(' '):
        if q1_word.strip().lower() in row['question2'].strip().lower():
            common_words_count+=1
    return common_words_count
```

```python
data['common_word_count'] = data.apply(find_common_words, axis=1)
data.head()
```
