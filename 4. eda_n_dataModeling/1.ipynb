## vectorization

```python
import pandas as pd

data = pd.read_csv(r'../1. data_gathering\data\train.csv\train.csv')
data.head()
```

```python
data.shape
```

```python
data = data.sample(30000, random_state=42)
```

```python
data.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

all_questions = data['question1'].tolist() + data['question2']
cv = CountVectorizer(max_features=3000)
cv.fit(all_questions)
```

```python
vectorized_question1 = pd.DataFrame(cv.transform(data['question1']).toarray())
vectorized_question2 = pd.DataFrame(cv.transform(data['question2']).toarray())
```

```python
vectorized_question1.shape, vectorized_question2.shape
```

```python
v_data = pd.concat([vectorized_question1, vectorized_question2], axis=1)
v_data
```

```python
data['is_duplicate']
```

```python
v_data.shape
```

```python
data = pd.concat([data['is_duplicate'].reset_index(drop=True), v_data.reset_index(drop=True)], axis=1)
```

```python
data['is_duplicate']
```

## data modeling starts


```python
from sklearn.model_selection import train_test_split

x = data.drop(columns=['is_duplicate'], axis=1)
y = data['is_duplicate']

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score

rf = RandomForestClassifier()
rf.fit(train_x, train_y)
test_y_pred = rf.predict(test_x)
accuracy_score(test_y, test_y_pred), precision_score(test_y, test_y_pred)
```
